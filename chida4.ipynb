【問題】栗まんじゅうが太陽系を覆う日

```python
# ふりかけられる物（まんじゅう）の体積をvolume　覆われる空間（太陽系）をspaceとする
# 覆われる時間をtime　とする

# まんじゅうの体積を0.001km^3とし、太陽系の体積を1.4e+18km^3とする
# 合わせてグラフも作成

volume=0.001
space=1.4e+18
chi=0
i=0
k=0
da=[]
ya=[]
while chi<=space :
    chi+=volume*(2**i)
    i+=1
    k+=5
    da.append(chi)
    ya.append(k) 

print((k-5),"分後に太陽系が栗まんじゅうで全て覆われる",sep="")
    
import matplotlib.pyplot as plt
plt.plot(ya,da)
plt.title("manju-")
plt.xlabel("time")
plt.ylabel("space")
plt.show()
```

汎用的な関数の作成

```python
# 関数名はchida
# 入力値はふりかけられる物の体積をvolume(km^3)　覆われる空間をspace(km^3)とし　出力値は覆われる時間とする
# ふりかけられた物は5分ごとに倍に増えていく
# 覆われる時間と合わせてグラフも出力する
# 出力例として野球ボール(0.005km^3)にバインバインをかけたときに、東京ドーム(1240km^3)が覆われる

def chida(volume,space):
    chi=0
    i=0
    k=0
    da=[]
    ya=[]    
    while chi<=space :
        chi+=volume*(2**i)
        i+=1
        k+=5
        da.append(chi)
        ya.append(k) 
    print((k-5),"分後に全て覆われる",sep="")
    import matplotlib.pyplot as plt
    plt.plot(ya,da)
    plt.title("baibain")
    plt.xlabel("time")
    plt.ylabel("space")
    plt.show()

chida(0.005,1240)
```
