【問題1】乱数の作成

```python
# 乱数を生成

import numpy as np
np.random.seed(0)
mean=(-3,0)
cov = [[1, 0.8], [0.8, 1]]
np.random.multivariate_normal(mean,cov,500)
```

【問題2】散布図による可視化

```python
# 作図

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mean=(-3,0)
cov = [[1, 0.8], [0.8, 1]]
x=np.random.multivariate_normal(mean,cov,500)[:,0]
y=np.random.multivariate_normal(mean,cov,500)[:,1]
plt.scatter(x, y)
```

【問題4】データの追加

```python
# 新たな乱数生成

import numpy as np
np.random.seed(0)
mean=(0,-3)
cov = [[1, 0.8], [0.8, 1]]
np.random.multivariate_normal(mean,cov,500)
```

```python
#　新たな乱数の作図

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
mean=(-3,0)
cov = [[1, 0.8], [0.8, 1]]
x=np.random.multivariate_normal(mean,cov,500)[:,0]
y=np.random.multivariate_normal(mean,cov,500)[:,1]
plt.scatter(x, y,label="0")


np.random.seed(0)
mean=(0,-3)
cov = [[1, 0.8], [0.8, 1]]
np.random.multivariate_normal(mean,cov,500)

x=np.random.multivariate_normal(mean,cov,500)[:,0]
y=np.random.multivariate_normal(mean,cov,500)[:,1]
plt.scatter(x, y,label="1",color="red")
plt.legend()
```

【問題5】データの結合

```python
# 結合と型の確認

import numpy as np

np.random.seed(0)
mean=(-3,0)
cov = [[1, 0.8], [0.8, 1]]
a1=np.random.multivariate_normal(mean,cov,500)


np.random.seed(0)
mean=(0,-3)
cov = [[1, 0.8], [0.8, 1]]
a2=np.random.multivariate_normal(mean,cov,500)


ya=np.concatenate([a1,a2])

ya.shape
```

【問題6】ラベル付け

```python
# ラベルをつける
# 問題５の結合データに00...0 1...11の列をさらに結合し(1000,3)の行列を作成

# 0が500個のリストを作る
chi=[]
for i in range(500):
    chi.append(0)

# 1が500個のリストを作る
da=[]
for i in range(500):
    da.append(1)
    
import numpy as np
chi1=np.array(chi)
da1=np.array(da)

# ラベル用の行が完成
su=np.concatenate([chi1,da1])

np.insert(ya,2,su,axis=1).shape
```
