【問題5】Pythonの関数化

```python
# 問題1から問題3を関数化

def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    import numpy as np
    array_x=np.arange(x_range[0],x_range[1],x_range[2])
    array_y=function(array_x)
    array_x1=np.matrix(array_x)
    array_y1=np.matrix(array_y)
    array_xy=np.insert(array_y1,0,array_x1,axis=0).T
    
    koubai1=[]
    for i in range(len(array_x)-1):
        if array_xy[i+1,0]==0:
            pass
        else:
            ya1=array_xy[i,1]/array_xy[i+1,0]
            ya2=round(ya1,2)
            koubai1.append(ya2)
    return koubai1

def function1(array_x):
    return (1/2)*array_x+1

x_range=(-50,50.1,0.1)
compute_gradient(function1,x_range)
```
