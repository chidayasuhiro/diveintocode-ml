【問題1】データの取得

```python
# データ取り込み
from sklearn.datasets import load_iris
iris=load_iris()
iris.data
```

```python
# PandasのDataFrame型に格納
import pandas as pd
import numpy as np

data1 = pd.DataFrame(iris.data,columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```

```python
data1
```

```python
# 目的変数の格納

target=[]
for i in range(150):
    target.append(0)
    
data2 = pd.DataFrame(target,columns=['Species'])
data2
```

【問題2】データの結合

```python
df=pd.concat([data1, data2], axis=1)
df
```

【問題3】データの確認

```python
# 先頭から4番目までのサンプルを表示させ、各特徴量がどういったものかを見る

df[0:4]
```

```python
# ラベルごとのサンプル数の合計を出力する
print(df["sepal_length"].sum())
print(df["sepal_width"].sum())
print(df["petal_length"].sum())
print(df["petal_width"].sum())
print(df["Species"].sum())
```

```python
# 特徴量に欠損値があるかを確認する

df.isnull().all()
```

```python
# 特徴量の平均値と標準偏差、四分位数を一度に表示させる
print(df.quantile([0.25,0.50,0.75]))
print(df.mean())
print(df.std())
```

【問題6】図の作成

```python
# ラベルごとのサンプル数を円グラフにする

import matplotlib.pyplot as plt
import numpy as np

# ラベルごとのサンプル数
sl=df["sepal_length"].sum()
sw=df["sepal_width"].sum()
pl=df["petal_length"].sum()
pw=df["petal_width"].sum()
sp=df["Species"].sum()

list_sample=[sl,sw,pl,pw,sp]

plt.pie(list_sample,autopct="%1.1f%%")
plt.show()
```

```python
# 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化

plt.boxplot(df["sepal_length"])
plt.show()
```

```python
plt.boxplot(df["petal_length"])
plt.show()
```
