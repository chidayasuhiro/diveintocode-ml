【問題1】100日目の米粒の数

```python
#　100日目までの合計もらっている数

rice=1
chi=0
for i in range(100):
    chi+=rice*(2**i)
print("合計： {}粒".format(chi))
```

```python
#　100日目までの「その日にもらう米の数」のグラフ

rice=1
chi=0
da=[]
for i in range(100) :
    chi=rice*(2**i)
    da.append(chi)
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("grain of rice")
plt.xlabel("day")
plt.ylabel("grain of rice")
plt.plot(da)
plt.show()
```

```python
#　100日目までの「その日までに累計で貰う米の数」のグラフ

rice=1
chi=0
da=[]
for i in range(100) :
    chi+=rice*(2**i)
    da.append(chi)
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("grain of rice")
plt.xlabel("day")
plt.ylabel("grain of rice")
plt.plot(da)
plt.show()
```

【問題3】貰える米で何人が何日生活できるか

```python
#　１人が最低限の生活ができる米粒の数＝12000粒（ネット調べ）
#　クラスの人数は15名
#　コース受講日数は123日

#　n日間でもらえる米粒の関数
def compute_sorori_shinzaemon(days): 
    rice=1
    chi=0
    for i in range(days):
        chi+=rice*(2**i)
    return chi

#　このクラスで何人生活できるか計算
print(compute_sorori_shinzaemon(123)//(15*12000),"日間生活できる",sep="")
```

```python
#　何人が何日生活できるか計算する関数

#　入力は 米粒の数(int)=rice と 人数(int)=x 出力は　生活できる日数(int)=life
#　関数名はchida
#　問題文の解釈が間違っていたら申し訳ないが、n日でも貰える米粒の数⇨米粒の数(int)=riceと解釈
#　関数出力例として、20日間で貰える量で10人が何日生活できるかを出す⇨chida(20,10)


def chida(days,x): 
    rice=1
    chi=0
    for i in range(days):
        chi+=rice*(2**i)
    life=chi//(x*12000)
    print(x,"人が秀吉から",days,"日分の褒美として貰える米粒で",chi//(x*12000),"日間生活できる",sep="")
    
chida(20,10)
```
